# Mini Prompt Engine using LangChain

## Objective
This project demonstrates how to replace hardcoded prompts with dynamic and reusable prompt templates using LangChain.

## Learning Outcomes
- Use PromptTemplate
- Build reusable prompt systems
- Handle multi-input prompts
- Implement validation logic
- Design prompt pipelines

```python
# Install langchain
!pip install langchain

from langchain_core.prompts import PromptTemplate
from langchain_core.prompts import ChatPromptTemplate
```

## Task 1: Replace Hardcoded Prompts

Convert a static hardcoded prompt into a reusable PromptTemplate.

```python
# Create reusable template
template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

def explain_topic(topic):
    return template.format(topic=topic)

# Output
print(explain_topic("Artificial Intelligence"))
```

## Task 2: Multi-Input Prompt System

This template supports multiple inputs: topic, audience, and tone.

```python
# Task 2: Multi-input prompt

multi_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

print(multi_template.format(topic="AI", audience="beginners", tone="friendly"))
print(multi_template.format(topic="Python", audience="kids", tone="fun"))
print(multi_template.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

## Task 3: Prompt Variations Engine

Different templates for teaching, interview, and storytelling styles.

```python
# Task 3: Prompt variations

teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

## Task 4: ChatPromptTemplate System

This task uses role-based chat prompts (teacher, interviewer, motivator).

```python
# Task 4: ChatPromptTemplate

def get_chat_prompt(role, topic):
    if role == "teacher":
        system_msg = "You are a helpful teacher."
    elif role == "interviewer":
        system_msg = "You are an interviewer."
    elif role == "motivator":
        system_msg = "You motivate learners."
    else:
        system_msg = "You are an assistant."

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_msg),
        ("user", "Explain {topic}")
    ])

    return chat_template.format_messages(topic=topic)

print(get_chat_prompt("teacher", "Neural Networks"))
```

## Task 5: Input Validation Layer

Ensures that audience and tone inputs are valid or assigns default values.

```python
# Task 5: Input validation

def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        audience = "beginner"

    if tone not in valid_tone:
        tone = "formal"

    return audience, tone

print(validate_inputs("kids", "cool"))
```

## Task 6: Prompt Generator App

Generates prompts dynamically based on style input.


```python
# Task 6: Prompt generator

def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)

    if style == "teaching":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} teaching style"
        )

    elif style == "interview":
        template = PromptTemplate(
            input_variables=["topic"],
            template="Ask interview questions about {topic}"
        )

    elif style == "storytelling":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} storytelling style"
        )

    else:
        raise ValueError("Invalid style")

    if style == "interview":
        return template.format(topic=topic)

    return template.format(topic=topic, audience=audience, tone=tone)

print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

## Task 7: Template Reusability Test

Same template used with multiple inputs to demonstrate reusability.

```python
# Task 7: Reusability

reusable_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms"
)

topics = ["AI", "Blockchain", "Cloud Computing", "Cyber Security", "Data Science"]

for t in topics:
    print(reusable_template.format(topic=t))
```

## Final Pipeline Flow

User Input → Validation → Prompt Template → Dynamic Prompt Generation → Output

```python
# Final pipeline

def mini_prompt_engine(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)
    return generate_prompt(topic, audience, tone, style)

print(mini_prompt_engine("Deep Learning", "beginner", "fun", "teaching"))
```
